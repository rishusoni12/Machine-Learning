# Problem Statement 

Farmers often face difficulty choosing the right crop for their land due to changing weather, soil conditions, and lack of expert advice. Wrong crop selection can lead to poor yield and financial loss. To solve this, we can use machine learning to predict the best crop to grow based on soil nutrients, temperature, humidity, pH level, and rainfall. This will help farmers make better decisions. It saves time, effort, and increases crop productivity.

Objective

The goal is to build a machine learning model that can recommend the most suitable crop for cultivation based on given environmental and soil conditions.

# Feature Engineering 

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df["label"] = le.fit_transform(df["label"])
```

```python
# To see label mappings
print(dict(zip(le.classes_, le.transform(le.classes_))))
```

# Model Building

```python
# Split features and target
X = df.drop(columns="label")
Y = df["label"]
```

```python
Y
```

```python
#Training Data 
from sklearn.model_selection  import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
Y_test
```

# Model selection Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
```

```python
dt.fit(X_train, Y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
#Measuring the model Accuracy 
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
print("Accuracy Score", accuracy_score(y_pred, Y_test))
print("classification_report",  classification_report(y_pred, Y_test))
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression(max_iter=1000, solver="liblinear")
```

```python
lr.fit(X_train, Y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
print("Accuracy Score", accuracy_score(y_pred, Y_test))
print("classification_report",  classification_report(y_pred, Y_test))
print("confusion_matrix", confusion_matrix(y_pred, Y_test))
```
