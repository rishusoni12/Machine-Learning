Cancer is one of the leading causes of death worldwide, and early detection is critical for improving survival rates.
Traditional methods of diagnosis, such as biopsies and imaging, are often invasive, time-consuming, and expensive.
With the increasing availability of healthcare data, there is an urgent need to develop automated, accurate, and efficient systems that can assist healthcare professionals in detecting and predicting cancer at an early stage.
The goal of this project is to develop a machine learning model that can accurately classify and predict cancer based on patient data. 
Predict if a breast tumor is cancerous or not using patient

Use labeled patient diagnostic data to train a machine learning classification model.
Predict whether a **breast tumor is malignant or benign** based on features such as radius, texture, perimeter, area, smoothness, etc.
